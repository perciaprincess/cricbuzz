```python
for type_match in data.get("typeMatches", []):
        for series_wrapper in type_match.get("seriesMatches", []):
            if "seriesAdWrapper" not in series_wrapper:
                continue
            series = series_wrapper["seriesAdWrapper"]
            series_id = series["seriesId"]
            series_name = series["seriesName"]

            for m in series["matches"]:
                match_info = m["matchInfo"]
                match_score = m.get("matchScore", {})

                # Detect winner
                status = match_info.get("status", "")
                winner_team = None
                winner_team_id = None
                if "won" in status.lower():
                    winner_team = status.split("won")[0].strip()
                    if winner_team.lower() in match_info["team1"]["teamName"].lower():
                        winner_team_id = match_info["team1"]["teamId"]
                    elif winner_team.lower() in match_info["team2"]["teamName"].lower():
                        winner_team_id = match_info["team2"]["teamId"]

                insert_query = """
                    insert into matches
                    (match_id, series_id, series_name, match_desc, match_format, start_date, end_date, 
                    state, status, team1_id, team1_name, team2_id, team2_name, venue_id, venue_name, 
                    venue_city, winner_team, winner_team_id)
                    VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)

                """
                insert_score_query = """
                    insert into match_scores
                    (match_id, team_id, team_name, innings_no, runs, wickets, overs)
                    VALUES (%s,%s,%s,%s,%s,%s,%s)
                """

                # Insert match info
                match_values = (
                    match_info["matchId"],
                    series_id,
                    series_name,
                    match_info.get("matchDesc"),
                    match_info.get("matchFormat"),
                    match_info.get("startDate"),
                    match_info.get("endDate"),
                    match_info.get("state"),
                    status,
                    match_info["team1"]["teamId"],
                    match_info["team1"]["teamName"],
                    match_info["team2"]["teamId"],
                    match_info["team2"]["teamName"],
                    match_info["venueInfo"]["id"],
                    match_info["venueInfo"]["ground"],
                    match_info["venueInfo"]["city"],
                    winner_team,
                    winner_team_id
                )
                cursor.execute(insert_query, match_values)

                # Insert scores (team1 innings, team2 innings)
                for team_key in ["team1", "team2"]:
                    score_data = match_score.get(f"{team_key}Score", {})
                    for inng_key, inng in score_data.items():
                        cursor.execute(
                            insert_score_query,
                            (
                                match_info["matchId"],
                                match_info[team_key]["teamId"],
                                match_info[team_key]["teamName"],
                                inng.get("inningsId"),
                                inng.get("runs"),
                                inng.get("wickets"),
                                inng.get("overs"),
                            )
                        )

# for type_block in data.get("typeMatches", []):
#     for series_block in type_block.get("seriesMatches", []):
#         series = series_block.get("seriesAdWrapper", {})
#         matches = series.get("matches", [])
        
#         for m in matches:
#             info = m.get("matchInfo", {})
#             venue = info.get("venueInfo", {})
            
#             match_id = info.get("matchId")
#             series_id = info.get("seriesId")
#             series_name = info.get("seriesName")
#             match_desc = info.get("matchDesc")
#             match_format = info.get("matchFormat")
#             state = info.get("state")
#             status = info.get("status")
#             team1 = info.get("team1", {}).get("teamName")
#             team2 = info.get("team2", {}).get("teamName")
#             venue_name = venue.get("ground")
#             city = venue.get("city")
#             start_date = info.get("startDate")
#             end_date = info.get("endDate")

#             query = """
#                 insert into matches
#                 (match_id, series_id, series_name, match_desc, match_format, state, status, 
#                  team1, team2, venue, city, start_date, end_date)
#                 VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
#             """
#             data = (match_id, series_id, series_name, match_desc, match_format, state, status,
#                   team1, team2, venue_name, city, start_date, end_date)
#             cursor.execute(query,data)
```
